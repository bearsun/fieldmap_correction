```python
def show_slices(slices):
    """Function to display row of image slices """ #ref: nypy.org/nibabel/coordinate_systems.html
    fig, axes = plt.subplots(1,len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap="gray", origin = "lower")
```


### Processing the fieldmap phase difference image

Phase values (0 to 360 degrees) in the PHASE difference image have been mapped to 0 to 4096 (Hz). The below step converts these values to radians per seconds (rad/s) by multiplying with 2pi. The value 2.46 is the echo time difference in ms.  More on this at **Registration_Practical_2015.pdf**.

The FSL tool, *fsl_prepare_fieldmap* is used here. The GUI version also can be used by typing *Fsl_prepare_fieldmap* in Lunix based system. 


```python
os.system('fsl_prepare_fieldmap SIEMENS FMAP_PHASE FMAP_MAG_brain_ero FMAP_RADS 2.46 ') #FMAP_MAG_brain_ero is used as the mask.
```
